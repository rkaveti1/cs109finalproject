Step 1:

Download stock price data from Yahoo 

```python
def load_SP500_list():
	url = "http://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
	page = web.DOM(requests.get(url).text)
	data = page.by_class('wikitable sortable')[0]
	columns = ['Symbol', 'Company', 'GICS', 'GICSSubIndustry', 'HQ', 'DateAdded']
	rows = data('tr')
	companies = []
	for i, r in enumerate(rows):
		if i > 0:
			cells = r('td')
			symbol = leaf_node(cells[0]).content
			company = leaf_node(cells[1]).content
			GICS = leaf_node(cells[3]).content
			GICSSub = leaf_node(cells[4]).content
			HQ = leaf_node(cells[5]).content
			date_added = None
			if len(cells) > 6:
				date_added = leaf_node(cells[6]).content
			companies.append([symbol, company, GICS, GICSSub, HQ, date_added])
	df = pd.DataFrame(companies, columns = columns).set_index('Symbol').sort_index()
	df.to_csv('data/SP500.txt')

def save_stock_data(symbol, period='daily'):
	try:
		url = "http://ichart.finance.yahoo.com/table.csv?s=" + symbol + "&ignore=.csv"
		if period == 'monthly':
			url += '&g=m'
		data = requests.get(url).text
		df = pd.read_csv(StringIO.StringIO(data), parse_dates=['Date'])
		df = df.set_index('Date').sort_index()
		returns = []
		prev = None
		for i, v in enumerate(df['Adj Close']):
			if i == 0:
				returns.append(0.0)
			else:
				returns.append(v / prev - 1)
			prev = v
		df['Return'] = pd.Series(returns, index=df.index)
		df.to_csv('data/%s/%s.csv' % (period, symbol))
	except Exception:
		print "Unable to download %s" % symbol

def load_symbols(exchange):
	return pd.read_csv('data/' + exchange + '.txt', sep='\t')

def save_all_stocks_data(period='daily'):
	for exchange in ['NASDAQ', 'NYSE']:
		for symbol in load_symbols(exchange)['Symbol']:
			save_stock_data(symbol, period)
			
def leaf_node(n):
	while n.children is not None and len(n.children) > 0 and n.children[0].type == 'element':
		n = n.children[0]
	return n
			
def save_analyst_opinion_data(symbol):
	try:
		url = "http://finance.yahoo.com/q/ud?s=" + symbol
		page = web.DOM(requests.get(url).text)
		data = page('table.yfnc_datamodoutline1 tr table tr')
		if len(data) > 1:
			columns = [leaf_node(n).content for n in data[0].children]
			entries = []
			for i in range(1, len(data)):
				entries.append([leaf_node(n).content for n in data[i].children])
			df = pd.DataFrame(entries, columns=columns)
			format = "%d-%b-%y"
			df['Date'] = pd.to_datetime(df['Date'], format=format)
			df = df.set_index('Date').sort_index(ascending=False)
			df.to_csv('data/analyst_opinion/' + symbol + '.txt')			
	except Exception:
		print "Unable to download analyst_opinion for %s" % symbol
		
def save_all_analyst_opinion_data():		
	for exchange in ['NASDAQ', 'NYSE']:
		for symbol in load_symbols(exchange)['Symbol']:
			save_analyst_opinion_data(symbol)
			
def save_all_stock_info():
	f = open('data/stock_info.csv', 'w')
	f.write("MarketCap,'PE','EPS','Dividend','DividentPct'\n")
	symbols, marketcaps, pes, epss, dividends, dividendpcts = [], [], [], [], [], []
	for exchange in ['NASDAQ', 'NYSE']:
		for symbol in load_symbols(exchange)['Symbol']:
			try:
				url = "http://finance.yahoo.com/q?s=" + symbol
				page = web.DOM(requests.get(url).text)
				data = page('table#table2 td.yfnc_tabledata1')
				market_cap = data[-4]('span')[0].content
				if market_cap[-1] == 'B':
					market_cap = float(market_cap[:-1]) * 10**9
				elif market_cap[-1] == 'M':
					market_cap = float(market_cap[:-1]) * 10**6
				else:
					market_cap = np.nan
				
				P_E = np.nan
				try:
					P_E = float(data[-3].content)
				except ValueError:
					P_E = np.nan
				
				EPS = np.nan
				try:
					EPS = float(data[-2].content)
				except ValueError:
					EPS = np.nan
				
				dividend_amount = np.nan
				dividend_pct = np.nan
				try:
					dividend_amount, dividend_pct = data[-1].content.strip().split(' ')
					dividend_amount = float(dividend_amount)
					dividend_pct = float(dividend_pct[1:-2])
				except ValueError:
					dividend_amount = np.nan
					dividend_pct = np.nan
				
				symbols.append(symbol)
				marketcaps.append(market_cap)
				pes.append(P_E)
				epss.append(EPS)
				dividends.append(dividend_amount)
				dividendpcts.append(dividend_pct)
			except Exception:
				print "Unable to get info for %s" % symbol
	df = pd.DataFrame({'Symbol': symbols, 'MarketCap': marketcaps, 'PE': pes, 'EPS': epss, 'Dividend': dividends, 'DividendPct': dividendpcts})
	df = df.set_index('Symbol').sort_index()
	df.to_csv('data/stock_info.txt')
	
def SP500_monthly_returns():
	from_date = dt.datetime(2012, 10, 1)
	to_date = dt.datetime(2013, 10, 1)
	sp500 = pd.read_csv('data/SP500.txt')['Symbol']
	df_returns = None
	for s in sp500:
		ret = pd.read_csv('data/monthly/%s.csv' % s, parse_dates=['Date']).set_index('Date')
		if ret.iloc[0].name > from_date:
			continue
		ret = ret[(ret.index >= from_date) & (ret.index <= to_date)]
		if df_returns is None:
			df_returns = pd.DataFrame({'Date': ret.index})
		ret = ret['Return']
		df_returns[s] = ret.values
	df_returns = df_returns.set_index('Date')
#	df_returns.to_csv('data/SP500_returns.txt')
	return df_returns

def save_indice_data(period='daily'):
	for s in ['^gspc', '^ixic']:
		save_stock_data(s, period)
```

Plot mean of monthly return vs. stdev of monthly return of SP500 stocks from Oct 2012 to Oct 2013

```python
def plot_SP500_monthly_returns():
#	df_returns = pd.read_csv('data/SP500_returns.txt').set_index('Date')
	df_returns = SP500_monthly_returns()
	mu = df_returns.mean(axis=0)
	sigma = df_returns.std(axis=0)
	plt.scatter(mu, sigma)
	plt.xlabel('Mean of monthly return')
	plt.ylabel('Std of monthly return')
	plt.title('S&P 500 components monthly return mean and std from Oct 2012 to Oct 2013')

plot_SP500_monthly_returns()
```

From Oct 2013 to Oct 2013, NFLX has stunning 16% average monthly return, but it also have 22% stdev.

```python
df_returns = SP500_monthly_returns()
mu = df_returns.mean(axis=0)
max_return = mu[mu == mu.max()]
std = df_returns[max_return.index[0]].std()
print "Max monthly return: %s %.2f stdev: %.2f" % (max_return.index[0], max_return[0], std)
```

Plot histogram for market capitalization (log base 10) for stocks in S&P 500

```python
df_stocks_info = pd.read_csv('data/stock_info.txt')
market_cap = df_stocks_info['MarketCap']
market_cap = np.log10(market_cap[~np.isnan(market_cap)])
mu = market_cap.mean()
sigma = market_cap.std()
plt.hist(market_cap, normed=True)
plt.xlabel('Market Capitalization')
plt.ylabel('Number of companies')
plt.title('S&P 500 components monthly return mean and std from Oct 2012 to Oct 2013')
plt.xticks(np.arange(6,13), ['1M', '10M', '100M', '1B', '10B', '100B', '1T'])
x = np.arange(6, 13, 0.1)
y = mlab.normpdf(x, mu, sigma)
plt.plot(x, y, 'r--')
```

Surprisingly, the distribution of market capitalization is very normal, centered around 1 billion dollars.
