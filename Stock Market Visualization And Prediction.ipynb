```python
def load_SP500_list():
	url = "http://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
	page = web.DOM(requests.get(url).text)
	data = page.by_class('wikitable sortable')[0]
	columns = ['Symbol', 'Company', 'GICS', 'GICSSubIndustry', 'HQ', 'DateAdded']
	rows = data('tr')
	companies = []
	for i, r in enumerate(rows):
		if i > 0:
			cells = r('td')
			symbol = leaf_node(cells[0]).content
			company = leaf_node(cells[1]).content
			GICS = leaf_node(cells[3]).content
			GICSSub = leaf_node(cells[4]).content
			HQ = leaf_node(cells[5]).content
			date_added = None
			if len(cells) > 6:
				date_added = leaf_node(cells[6]).content
			companies.append([symbol, company, GICS, GICSSub, HQ, date_added])
	df = pd.DataFrame(companies, columns = columns).set_index('Symbol').sort_index()
	df.to_csv('data/SP500.txt')

def save_stock_data(symbol, period='daily'):
	try:
		url = "http://ichart.finance.yahoo.com/table.csv?s=" + symbol + "&ignore=.csv"
		if period == 'monthly':
			url += '&g=m'
		data = requests.get(url).text
		df = pd.read_csv(StringIO.StringIO(data), parse_dates=['Date'])
		df = df.set_index('Date').sort_index()
		returns = []
		prev = None
		for i, v in enumerate(df['Adj Close']):
			if i == 0:
				returns.append(0.0)
			else:
				returns.append(v / prev - 1)
			prev = v
		df['Return'] = pd.Series(returns, index=df.index)
		df.to_csv('data/%s/%s.csv' % (period, symbol))
	except Exception:
		print "Unable to download %s" % symbol

def load_symbols(exchange):
	return pd.read_csv('data/' + exchange + '.txt', sep='\t')

def save_all_stocks_data(period='daily'):
	for exchange in ['NASDAQ', 'NYSE']:
		for symbol in load_symbols(exchange)['Symbol']:
			save_stock_data(symbol, period)
			
def leaf_node(n):
	while n.children is not None and len(n.children) > 0 and n.children[0].type == 'element':
		n = n.children[0]
	return n
			
def save_analyst_opinion_data(symbol):
	try:
		url = "http://finance.yahoo.com/q/ud?s=" + symbol
		page = web.DOM(requests.get(url).text)
		data = page('table.yfnc_datamodoutline1 tr table tr')
		if len(data) > 1:
			columns = [leaf_node(n).content for n in data[0].children]
			entries = []
			for i in range(1, len(data)):
				entries.append([leaf_node(n).content for n in data[i].children])
			df = pd.DataFrame(entries, columns=columns)
			format = "%d-%b-%y"
			df['Date'] = pd.to_datetime(df['Date'], format=format)
			df = df.set_index('Date').sort_index(ascending=False)
			df.to_csv('data/analyst_opinion/' + symbol + '.txt')			
	except Exception:
		print "Unable to download analyst_opinion for %s" % symbol
		
def save_all_analyst_opinion_data():		
	for exchange in ['NASDAQ', 'NYSE']:
		for symbol in load_symbols(exchange)['Symbol']:
			save_analyst_opinion_data(symbol)
			
def save_all_stock_info():
	f = open('data/stock_info.csv', 'w')
	f.write("MarketCap,'PE','EPS','Dividend','DividentPct'\n")
	symbols, marketcaps, pes, epss, dividends, dividendpcts = [], [], [], [], [], []
	for exchange in ['NASDAQ', 'NYSE']:
		for symbol in load_symbols(exchange)['Symbol']:
			try:
				url = "http://finance.yahoo.com/q?s=" + symbol
				page = web.DOM(requests.get(url).text)
				data = page('table#table2 td.yfnc_tabledata1')
				market_cap = data[-4]('span')[0].content
				if market_cap[-1] == 'B':
					market_cap = float(market_cap[:-1]) * 10**9
				elif market_cap[-1] == 'M':
					market_cap = float(market_cap[:-1]) * 10**6
				else:
					market_cap = np.nan
				
				P_E = np.nan
				try:
					P_E = float(data[-3].content)
				except ValueError:
					P_E = np.nan
				
				EPS = np.nan
				try:
					EPS = float(data[-2].content)
				except ValueError:
					EPS = np.nan
				
				dividend_amount = np.nan
				dividend_pct = np.nan
				try:
					dividend_amount, dividend_pct = data[-1].content.strip().split(' ')
					dividend_amount = float(dividend_amount)
					dividend_pct = float(dividend_pct[1:-2])
				except ValueError:
					dividend_amount = np.nan
					dividend_pct = np.nan
				
				symbols.append(symbol)
				marketcaps.append(market_cap)
				pes.append(P_E)
				epss.append(EPS)
				dividends.append(dividend_amount)
				dividendpcts.append(dividend_pct)
			except Exception:
				print "Unable to get info for %s" % symbol
	df = pd.DataFrame({'Symbol': symbols, 'MarketCap': marketcaps, 'PE': pes, 'EPS': epss, 'Dividend': dividends, 'DividendPct': dividendpcts})
	df = df.set_index('Symbol').sort_index()
	df.to_csv('data/stock_info.txt')
	
def SP500_monthly_returns():
	from_date = dt.datetime(2012, 10, 1)
	to_date = dt.datetime(2013, 10, 1)
	sp500 = pd.read_csv('data/SP500.txt')['Symbol']
	df_returns = None
	for s in sp500:
		ret = pd.read_csv('data/monthly/%s.csv' % s, parse_dates=['Date']).set_index('Date')
		if ret.iloc[0].name > from_date:
			continue
		ret = ret[(ret.index >= from_date) & (ret.index <= to_date)]
		if df_returns is None:
			df_returns = pd.DataFrame({'Date': ret.index})
		ret = ret['Return']
		df_returns[s] = ret.values
	df_returns = df_returns.set_index('Date')
#	df_returns.to_csv('data/SP500_returns.txt')
	return df_returns

def save_indice_data(period='daily'):
	for s in ['^gspc', '^ixic']:
		save_stock_data(s, period)
```

```python
import os.path

def create_master_df():
    
    zero_volume_thresh=30 #if more than this many days with zero volume, then don't include
    suspended_threshold=3 #if a stock has a return of 0 for threshold consecutive days, its probably suspended or it's very illiquid and we wont include it
    initial_date = dt.datetime(2007, 11, 1) #we need to calculate 200 SMA, so we need data before 2008
    from_date = dt.datetime(2008, 10, 1)
    to_date = dt.datetime(2013, 10, 1)
	
    ret = pd.read_csv('data/daily/^gspc.csv', parse_dates=['Date']).set_index('Date')
    closes = ret[(ret.index >= initial_date) & (ret.index <= to_date)]
    ret = ret[(ret.index >= from_date) & (ret.index <= to_date)]
    df_returns = pd.DataFrame({'Date': ret.index})
    df_returns['^gspc'] = ret['Return'].values
    df_returns = df_returns.set_index('Date')
    df_volume = pd.DataFrame({'Date': closes.index})
    df_volume['^gspc'] = closes['Volume'].values
    df_volume = df_volume.set_index('Date')
    df_closes = pd.DataFrame({'Date': closes.index})
    df_closes['^gspc'] = closes['Close'].values
    df_closes = df_closes.set_index('Date')
    
    df_stocks_info = pd.read_csv('data/stock_info.txt')
    df_stocks_info = df_stocks_info[~np.isnan(df_stocks_info['MarketCap'])]
    good_stocks = df_stocks_info['Symbol'].values
    
    for exchange in ['NASDAQ', 'NYSE']:
        for stock in load_symbols(exchange)['Symbol']:
            good_stock=1
            if stock not in good_stocks:
                continue
            if not os.path.isfile('data/daily/%s.csv' % stock):
                continue
            ret = pd.read_csv('data/daily/%s.csv' % stock, parse_dates=['Date']).set_index('Date')
            if ret.iloc[0].name > initial_date:
                continue
            closes = ret[(ret.index >= initial_date) & (ret.index <= to_date)]
            if len(closes)!=len(df_closes):
                continue
            ret = ret[(ret.index >= from_date) & (ret.index <= to_date)]
            if len(ret)!=len(df_returns):
                continue
            ret=ret['Return'].values
            vol=closes['Volume'].values
            close=closes['Close'].values
            zero_volume_index=np.where(vol[(closes.index >= from_date) & (closes.index <= to_date)]==0)[0]
            if len(zero_volume_index) > zero_volume_thresh:
                continue
            zero_index=np.where(ret==0)[0]
            if len(zero_index) > suspended_threshold:
                for i,z in enumerate(zero_index[0:-2]):
                    test_range=range(i+1,i+suspended_threshold+1)
                    if sum([j in zero_index for j in test_range])==len(test_range):
                        good_stock=0
                        continue
            
            if good_stock:
                df_returns[stock] = ret
                df_volume[stock] = vol
                df_closes[stock] = close
    
    testing_start = dt.datetime(2011, 10, 1)
    df_returns_train=df_returns[df_returns.index<testing_start]
    df_returns_test=df_returns[df_returns.index>=testing_start]
    df_returns_train.to_csv('data/master_returns_train.txt')
    df_returns_test.to_csv('data/master_returns_test.txt')
    
    df_volume.to_csv('data/factors/volume.txt')
    
    df_closes.to_csv('data/factors/close_prices.txt')

    return df_returns_train,df_returns_test
    
    
```

```python
from scipy import stats
from collections import defaultdict

def calculate_betas_ols(df_returns,hedge='^gspc'):
    hedge_returns=df_returns[hedge].values
    betas=defaultdict(float)
    alphas=defaultdict(float)
    for symbol in df_returns.columns:
        if symbol==hedge:
            continue        
        data = stats.linregress(hedge_returns,df_returns[symbol].values)
        betas[symbol]=data[0] 
        alphas[symbol]=data[1]
    return betas,alphas
    
```

```python
from collections import defaultdict
import statsmodels.api as sm
import sys

def calculate_betas_rlm(df_returns, M,hedge='^gspc'):
    hedge_returns=df_returns[hedge].values
    hedge_returns=sm.add_constant(hedge_returns,prepend=True)
    betas=defaultdict(float)
    alphas=defaultdict(float)
    for symbol in df_returns.columns:
        if symbol==hedge:
            continue        
        rlm_results=sm.RLM(df_returns[symbol].values,hedge_returns,M).fit()
        betas[symbol]=rlm_results.params[1]
        alphas[symbol]=rlm_results.params[0]
        
    return betas,alphas
```

Obtain monthly macro economics data from Fed Reserv. Data were pre-downloaded to data/macro folder from Fed Reserv (http://research.stlouisfed.org/fred2/). Each function (for example: get_CPI_monthly) reads in the csv file, and returns a data frame indexed by the first day of the month (datetime object) and having two columns: 'VALUE': the value in that month, 'ChangePct': percentage change from previous month. 

The macro economics features and the corresponding functions to return the data frame are specified in macro_features dictionary.


```python
def filter_index(df, start_index=None, end_index=None):    
    df1 = df.copy()
    if start_index is not None:
        df1 = df1[df1.index >= start_index]
    if end_index is not None:
        df1 = df1[df1.index <= end_index]
    return df1

def add_ratio(df, column='Value', ratio_column='ChangePct'):
    prev_value = None; changes = []
    for v in df[column]:
        if prev_value is None:
            changes.append(0.0)
        else:
            changes.append(v / prev_value - 1)
        prev_value = v
    df[ratio_column] = changes
    return df

def read_fed_serv_data(file, start_date, end_date):        
    df_full = pd.read_csv('data/macro/%s.csv' % file, parse_dates=['DATE'])
    df_full.rename(columns={'DATE':'Date', 'VALUE': 'Value'}, inplace=True)
    df = df_full.set_index('Date')
    df = filter_index(df, start_date, end_date)
    df = add_ratio(df)
    return df

def get_GDP_growth_monthly(start_date=None, end_date=None):
    df_full = pd.read_csv('data/macro/GDPC1.csv', parse_dates=['DATE']).set_index('DATE')
    df = df_full
    dates = []; values = []
    prev_quarterly = None
    for d, v in df.iterrows():
        if prev_quarterly is not None:
            dates.append(d + relativedelta(months=1))
            dates.append(d + relativedelta(months=2))
            values.append((v['VALUE'] - prev_quarterly) / 3 + prev_quarterly)
            values.append((v['VALUE'] - prev_quarterly) * 2 / 3 + prev_quarterly)
        prev_quarterly = v['VALUE']
        dates.append(d + relativedelta(months=3))
        values.append(v['VALUE'])
    df1 = pd.DataFrame({'Date': dates, 'Value': values}).set_index('Date')
    df1 = filter_index(df1, start_date, end_date)
    df1 = add_ratio(df1)
    return df1

def get_CPI_monthly(start_date=None, end_date=None):
    return read_fed_serv_data('CPIAUCSL', start_date, end_date)           

def get_industry_production_index_monthly(start_date=None, end_date=None):
    return read_fed_serv_data('INDPRO', start_date, end_date)          
    
def get_10_year_treasury_rate_monthly(start_date=None, end_date=None):
    return read_fed_serv_data('DGS10', start_date, end_date)          

def get_US_dollar_index_monthly(start_date=None, end_date=None):
    return read_fed_serv_data('TWEXMMTH', start_date, end_date)          

def get_payroll_employment_monthly(start_date=None, end_date=None):
    return read_fed_serv_data('PAYEMS', start_date, end_date)          

def get_unemployment_rate_monthly(start_date=None, end_date=None):
    return read_fed_serv_data('UNRATE', start_date, end_date)          

def get_m2_monthly(start_date=None, end_date=None):
    return read_fed_serv_data('M2', start_date, end_date)          
        
def get_p_e_ratio_monthly(start_date=None, end_date=None):
    path = 'data/micro/pe.csv'
    if not os.path.isfile(path):
        page = web.DOM(requests.get('http://www.multpl.com/table?f=m').text)
        rows = page('#datatable tr')
        dates = []; values = []
        for r in rows[1:]:
            dates.append(dt.datetime.strptime(r('td')[0].content, '%b %d, %Y'))
            values.append(float(r('td')[1].content.split(' ')[0]))
        df = pd.DataFrame({'Date': dates, 'Value': values}).set_index('Date').sort_index()
        df.to_csv(path)
    
    df = pd.read_csv(path, parse_dates=['Date']).set_index('Date')
    df = filter_index(df, start_date, end_date)
    df = add_ratio(df)
    return df

macro_features = {
        'GDP': get_GDP_growth_monthly,
        'CPI': get_CPI_monthly,
        'IP': get_industry_production_index_monthly,
        'Treasury': get_10_year_treasury_rate_monthly,
        'Dollar': get_US_dollar_index_monthly,
        'Employment': get_payroll_employment_monthly,
        'Unemployment': get_unemployment_rate_monthly,
        'M2': get_m2_monthly
    }
```

```python
def calculate_SMA(prices,time_period):
    data=np.zeros(len(prices)-time_period+1)
    for i in range(len(data)):
        data[i]=np.mean(prices[i:(i+time_period)])
    return data
```

```python
def calculate_EMA(prices,time_period):
    data=np.zeros(len(prices)-time_period+1)
    data[0]=np.mean(prices[0:time_period])
    mult=2./(float(time_period)+1.)
    for i in range(1,len(data)):
        data[i]=(prices[time_period+i-1]-data[i-1])*mult+data[i-1]
    return data
```

```python
def calculate_moving_sd(prices,time_period):
    data=np.zeros(len(prices)-time_period+1)
    for i in range(len(data)):
        data[i]=np.std(prices[i:(i+time_period)])
    return data
```

```python
def calculate_bollinger_band_values(prices,time_period=20):
    sma=calculate_SMA(prices,time_period)
    sd=calculate_moving_sd(prices,time_period)
    prices=prices[time_period-1:]
    return (prices-sma)/sd
    
```

```python
def calculate_MACD(prices,m_time_period1=12,m_time_period2=26,signal_time_period=9):
    ema1=calculate_EMA(prices,m_time_period1)
    ema2=calculate_EMA(prices,m_time_period2)
    ema1=ema1[-len(ema2):]
    macd=ema1-ema2
    s_ema=calculate_EMA(ema2,signal_time_period)
    macd=macd[-len(s_ema):]
    return macd-s_ema
    
```

```python
def calculate_volume_zscore(volumes,time_period=20):
    sma=calculate_SMA(volumes,time_period)
    sd=calculate_moving_sd(volumes,time_period)
    volumes=volumes[-len(sd):]
    data=(volumes-sma)/sd
    data[np.isnan(data)]=0
    return data
    
```

```python
def calculate_ema_crosses(prices,time_period1=50,time_period2=200):
    sma1=calculate_SMA(prices,time_period1)
    sma2=calculate_SMA(prices,time_period2)
    sma1=sma1[-len(sma2):]
    return sma1-sma2
```

```python
def calculate_williamsR(prices,time_period=14):
    data=np.zeros(len(prices)-time_period+1)
    for i in range(len(data)):
        highest_high=max(prices[i:(i+time_period-1)])
        lowest_low=min(prices[i:(i+time_period-1)])
        data[i]=-100.*(highest_high-prices[i+time_period-1])/(highest_high-lowest_low)
    return data
```

```python
def calculate_RSI(prices,time_period=14):
    returns=(prices[1:]-prices[:-1])/prices[:-1]
    data=np.zeros(len(returns)-time_period+1)
    for i in range(0,len(data)):
        gains_index=np.where(returns[i:(time_period-1+i)]>0)[0]
        if len(gains_index):
            gains=sum(returns[gains_index])
        else:
            gains=0
        loss_index=np.where(returns[i:(time_period-1+i)]<0)[0]
        if len(loss_index):
            loss=sum(returns[loss_index])
            data[i]=100.-100./(1+gains/loss)
        else:
            if gains==0:
                data[i]=50
            else:
                data[i]=100
    return data
```

```python
def make_factor_df(factor_name,symbols,train_dates,test_dates,data_df):
    
    factor_train_df=pd.DataFrame({'Date': train_dates})
    factor_test_df=pd.DataFrame({'Date': test_dates})
    
    if factor_name=='bollinger_bands':
        temp=calculate_bollinger_band_values(data_df[symbols[0]].values)
        temp_dates=data_df.index[-len(temp):]
        train_index=[np.where(temp_dates==d)[0] for d in train_dates]
        test_index=[np.where(temp_dates==d)[0] for d in test_dates]
        for s in symbols:
            vals=calculate_bollinger_band_values(data_df[s].values)
            factor_train_df[s]=vals[train_index]
            factor_test_df[s]=vals[test_index]
    elif factor_name=='macd':
        temp=calculate_MACD(data_df[symbols[0]].values)
        temp_dates=data_df.index[-len(temp):]
        train_index=[np.where(temp_dates==d)[0] for d in train_dates]
        test_index=[np.where(temp_dates==d)[0] for d in test_dates]
        for s in symbols:
            vals=calculate_MACD(data_df[s].values)
            factor_train_df[s]=vals[train_index]
            factor_test_df[s]=vals[test_index]
    elif factor_name=='volume_zscore':
        temp=calculate_volume_zscore(data_df[symbols[0]].values)
        temp_dates=data_df.index[-len(temp):]
        train_index=[np.where(temp_dates==d)[0] for d in train_dates]
        test_index=[np.where(temp_dates==d)[0] for d in test_dates]
        for s in symbols:
            vals=calculate_volume_zscore(data_df[s].values)
            factor_train_df[s]=vals[train_index]
            factor_test_df[s]=vals[test_index] 
    elif factor_name=='ema_crosses':
        temp=calculate_ema_crosses(data_df[symbols[0]].values)
        temp_dates=data_df.index[-len(temp):]
        train_index=[np.where(temp_dates==d)[0] for d in train_dates]
        test_index=[np.where(temp_dates==d)[0] for d in test_dates]
        for s in symbols:
            vals=calculate_ema_crosses(data_df[s].values)
            factor_train_df[s]=vals[train_index]
            factor_test_df[s]=vals[test_index]
    elif factor_name=='williams_R':
        temp=calculate_williamsR(data_df[symbols[0]].values)
        temp_dates=data_df.index[-len(temp):]
        train_index=[np.where(temp_dates==d)[0] for d in train_dates]
        test_index=[np.where(temp_dates==d)[0] for d in test_dates]
        for s in symbols:
            vals=calculate_williamsR(data_df[s].values)
            factor_train_df[s]=vals[train_index]
            factor_test_df[s]=vals[test_index]
    elif factor_name=='RSI':
        temp=calculate_RSI(data_df[symbols[0]].values)
        temp_dates=data_df.index[-len(temp):]
        train_index=[np.where(temp_dates==d)[0] for d in train_dates]
        test_index=[np.where(temp_dates==d)[0] for d in test_dates]
        for s in symbols:
            vals=calculate_RSI(data_df[s].values)
            factor_train_df[s]=vals[train_index]
            factor_test_df[s]=vals[test_index]        
        
    factor_train_df.to_csv('data/factors/' + factor_name + '_train.txt')
    factor_test_df.to_csv('data/factors/' + factor_name + '_test.txt')
        
        
        
```

```python
#given data over dates, return the dates where the data is above (or below) the threshold for n_days in a row

def get_interest_dates(data,dates,threshold,n_days,greater_than_bool):
    if greater_than_bool:
        index=np.where(data>threshold)
    else:
        index=np.where(data<threshold)
    if len(index)>0:
        if n_days==1:
            return dates[index]
        else:
            dates=[]
            for i in range(len(index)-n_days+1):
                if index[i:(i+n_days-1)] == range(i,i+ndays):
                    dates.append(index[i])
            return dates
        
```
